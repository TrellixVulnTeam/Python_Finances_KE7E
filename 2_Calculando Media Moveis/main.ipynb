Obtendo os dados do Mercado

```python
ibov = web.get_data_yahoo('^BVSP', start='2010-05-03', end='2013-03-01')
```

```python
ibov.head()
```

```python
ibov.tail()
```

```python
ibov["Close"].plot(figsize=(22,8), label="IBOV")
ibov["Close"].rolling(21).mean().plot(label="MM21")
ibov["Close"].rolling(200).mean().plot(label="MM200")
plt.legend()
```

```python
ibov_fatiado = ibov[ibov.index.year == 2012]
ibov_fatiado["Close"].plot(figsize=(22,8), label="IBOV")
ibov_fatiado["Close"].rolling(21).mean().plot(label="MM21")
ibov_fatiado["Close"].rolling(200).mean().plot(label="MM200")
plt.legend();
```

```python
ibov_fatiado_2 = ibov[(ibov.index.year >= 2008) & (ibov.index.year <= 2015)]
ibov_fatiado_2["Close"].plot(figsize=(22,8), label="IBOV")
ibov_fatiado_2["Close"].rolling(21).mean().plot(label="MM21")
ibov_fatiado_2["Close"].rolling(200).mean().plot(label="MM200")
plt.legend();
```
